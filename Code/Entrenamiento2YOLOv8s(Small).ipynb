```python
#!pip install ultralytics
from ultralytics import YOLO
# Cargar modelo base (YOLOv8s)
model = YOLO('yolov8s.pt')

# Entrenar el modelo
model.train(
    data='/content/drive/MyDrive/Colab Notebooks/TFM/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='visdrone-person-yolov8s'
)
```

```python
from ultralytics import YOLO
import os
import cv2
from tqdm import tqdm

# Rutas
model_path = '/content/drive/MyDrive/Colab Notebooks/TFM/YOLOV8S Compilation/visdrone-person-yolov8s/weights/best.pt'
video_input_path = '/content/drive/MyDrive/Colab Notebooks/TFM/VisDrone2019-YOLO/IMG_TEST_MANUAL/furbo_short.mp4'
video_output_path = '/content/drive/MyDrive/Colab Notebooks/TFM/YOLOV8S Compilation/VideoResultados/resultado_furbo_short.mp4'

# Crear directorio si no existe
os.makedirs(os.path.dirname(video_output_path), exist_ok=True)

# Cargar modelo
model = YOLO(model_path)

# Abrir video de entrada
cap = cv2.VideoCapture(video_input_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Crear escritor de video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(video_output_path, fourcc, fps, (width, height))

# Procesar frame a frame
for _ in tqdm(range(total_frames), desc="Procesando video"):
    ret, frame = cap.read()
    if not ret:
        break
    results = model.predict(frame, verbose=False)
    annotated_frame = results[0].plot()
    out.write(annotated_frame)

# Liberar recursos
cap.release()
out.release()

print(f"âœ… Video anotado guardado en: {video_output_path}")
```
